```python
from googleapiclient.discovery import build
from google_auth_oauthlib.flow import InstalledAppFlow
import os
# Scopes: this one allows YouTube playlist modification
SCOPES = ["https://www.googleapis.com/auth/youtube"]

# Authenticate and get API client
flow = InstalledAppFlow.from_client_secrets_file(
    "client_secret.json", SCOPES
)
credentials = flow.run_local_server(port=0)

youtube = build("youtube", "v3", credentials=credentials)
# Step 1: Find Cinedesi channel ID
search_channel = youtube.search().list(
    q="cinedesi",
    type="channel",
    part="snippet",
    maxResults=1
).execute()

channel_id = search_channel['items'][0]['snippet']['channelId']

# Step 2: Search for GOT videos
videos = []
next_page_token = None
while True:
    search_videos = youtube.search().list(
        channelId=channel_id,
        q="GAME OF THRONES",
        type="video",
        part="snippet",
        maxResults=50,
        pageToken=next_page_token
    ).execute()

    for item in search_videos['items']:
        videos.append(item['id']['videoId'])

    next_page_token = search_videos.get('nextPageToken')
    if not next_page_token:
        break
```

```python
# Step 3: Create playlist
playlist_request = youtube.playlists().insert(
    part="snippet,status",
    body={
        "snippet": {
            "title": "Cinedesi GOT Reactions",
            "description": "All Game of Thrones reaction videos by Cinedesi",
            "tags": ["Game of Thrones", "Cinedesi", "Reaction"]
        },
        "status": {
            "privacyStatus": "private"
        }
    }
).execute()

playlist_id = playlist_request["id"]
```

```python
import json
import os

# Quota-safe parameters
MAX_VIDEOS_PER_RUN = 50  # change as needed

# Load unfinished list if resuming
if os.path.exists("remaining_videos.json"):
    with open("remaining_videos.json", "r") as f:
        videos = json.load(f)

# Determine how many videos to process today
videos_to_add = videos[:MAX_VIDEOS_PER_RUN]
remaining_videos = videos[MAX_VIDEOS_PER_RUN:]

# Add videos to playlist
for vid in videos_to_add:
    youtube.playlistItems().insert(
        part="snippet",
        body={
            "snippet": {
                "playlistId": playlist_id,
                "resourceId": {
                    "kind": "youtube#video",
                    "videoId": vid
                }
            }
        }
    ).execute()
    print(f"Added video: {vid}")

# Save remaining videos for tomorrow
with open("remaining_videos.json", "w") as f:
    json.dump(remaining_videos, f)

print(f"Playlist created/updated: https://www.youtube.com/playlist?list={playlist_id}")
print(f"{len(videos_to_add)} videos added today. {len(remaining_videos)} still remaining.")
```
